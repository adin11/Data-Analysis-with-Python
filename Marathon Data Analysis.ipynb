## 2. What age groups are best in 50m race (20 + races min)

```python
df2.query('`Event distance/length` == "50mi"') \
    .groupby(['athlete_age'])['Athlete average speed'] \
    .agg(['mean','count']) \
    .reset_index() \
    .sort_values('mean', ascending=False) \
    .query('count > 19')
```

So, overall, this code filters data for the '50mi' event distance, calculates the mean and count of 'Athlete average speed' for each age group, sorts the results by mean speed, and filters out groups with less than 20 data points.

## 2. Seasons for the data => Slower in the summer than in winter ?

#Spring 3-5
#Summer 6-8
#Fall 9-11
#Winter 12-2

```python
# making new column which consists of months  
df.head(2)
```

```python
df2['month'] = df2['Event dates'].str.split('.').str.get(1).astype(int)
```

```python
df2.head(3)
```

```python
df2['month'].unique() # Made these values positive 
```

```python
df2['month'] = df2['month'].abs()
```

```python
df2['month'].unique()
```

```python
# Here we hve created a new column based on the month column ,which will consist of the season 

df2['race_season'] = df2['month'].apply(lambda x: 'Spring' if x >= 3 and x <= 5 else
                                           'Summer' if x >= 6 and x <= 8 else
                                           'Fall' if x >= 9 and x <= 11 else
                                           'Winter')
```

```python
df2.head(10)
```

```python
df2['race_season'].unique()
```

### 2 . Seasons for the data => slower in the summer than in the winter ? 

```python
df2.groupby(['race_season'])['Athlete average speed'].agg(['mean','count']) \
 .sort_values('mean',ascending = True) \
 .reset_index()
```

```python
df2.groupby(['race_season'])['Athlete average speed'].agg(['mean','count']) \
 .sort_values('mean',ascending = True) \
.plot(kind = 'barh',title='Count of runners acc to seasons')
  
```
