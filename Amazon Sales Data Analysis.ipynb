# Performing EDA

```python
df.head(3)
```

```python
sns.countplot(data =df,x='Size')
```

```python
ax = sns.countplot(data=df,x='Size')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
size_data = df.groupby(['Size'],as_index = False)['Qty'].sum().sort_values(by='Qty',ascending=False)
sns.barplot(data=size_data,x='Size',y='Qty')
```

Note: From above Graph you can see that most of the Qty buys M-Size in the sales

```python
ax = sns.countplot(data=df,x='Courier Status', hue='Status')
sns.set(rc={'figure.figsize':(20,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

Note : From the Above Graph We can see that Most of the Orders are shipped and Delivered to the Buyer

```python
t_shirtdata = df.groupby(['Category'],as_index = False)['Qty'].sum().sort_values(by='Qty',ascending = False)
sns.barplot(data=t_shirtdata,x='Category',y='Qty')
```

Note : From This we can Understand that most of the people have purchased T-Shirts

```python
ax = sns.countplot(data=df,x='B2B')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
b2b_data = df.groupby(['B2B'],as_index=False)['Qty'].sum().sort_values(by='Qty' , ascending = False)
sns.barplot(data=b2b_data,x='B2B',y='Qty')
```

Note : From above chart we can see that maximum i.e. 99.2% of buyers are retailers and 0.7% are B2B buyers
